```python
from recipereg import *
from recipe import *
from currency import *
from disenchant import *
from id_to_mat import *
from item import *
from profitfinder import *
from tbcgear import *
from tradegoods import *
from lookup import *
from util import *
from main import *
```

```python
directory = "../data/"
auction = get_auction("D:/World of Warcraft/_classic_/WTF/Account/308707580#1/SavedVariables/JSON/data.json")
update_essences(auction)

mine = json.load(open(directory+"tbc_mining.json"))
cook = json.load(open(directory+"tbc_cooking.json"))
alch = json.load(open(directory+"tbc_alchemy.json"))
ench = json.load(open(directory+"tbc_enchants.json"))
tail = json.load(open(directory+"tbc_tailoring.json"))
leath = json.load(open(directory+"tbc_leatherworking.json"))
black = json.load(open(directory+"tbc_blacksmithing.json"))
jewel = json.load(open(directory+"tbc_jewelcrafting.json"))
engin = json.load(open(directory+"tbc_engineering.json"))
first = json.load(open(directory+"tbc_first_aid.json"))

tbc_gear = TBCGear()
tbc_gear.parse(directory+"tbc_armor.json", ItemClass.ARMOR)
tbc_gear.parse(directory+"tbc_weapons.json", ItemClass.WEAPON)

ID_to_Mat.parse(ench)  # gets dust IDs for disenchant only
ID_to_Mat.parse(tail)

Disenchant.registerAuction(auction)
Disenchant.registerIDtoName(ID_to_Mat.id_to_mat)
Disenchant.registerGear(tbc_gear)

Util.register(Disenchant, tbc_gear)

trade_goods = TradeGoods(directory+"tbc_trade_goods.json")
trade_goods.update_auction(auction)

cooking = RecipeRegistry()
cooking.recipes = RecipeBridge.parse(cook)
cooking.update_matCosts(auction)
cooking.update_craftable_matCosts()

firstaid = RecipeRegistry()
firstaid.recipes = RecipeBridge.parse(first)
firstaid.update_matCosts(auction)
firstaid.update_craftable_matCosts()

enchanting = RecipeRegistry()
enchanting.recipes = RecipeBridge.parse(ench)
enchanting.update_matCosts(auction)
enchanting.update_craftable_matCosts()

tailoring = RecipeRegistry()
tailoring.recipes = RecipeBridge.parse(tail)
tailoring.update_matCosts(auction)
tailoring.update_craftable_matCosts()

leatherworking = RecipeRegistry()
leatherworking.recipes = RecipeBridge.parse(leath)
leatherworking.update_matCosts(auction)
leatherworking.update_craftable_matCosts()

alchemy = RecipeRegistry()
alchemy.recipes = RecipeBridge.parse(alch)
alchemy.update_matCosts(auction)
alchemy.update_craftable_matCosts()

blacksmithing = RecipeRegistry()
blacksmithing.recipes = RecipeBridge.parse(black)
blacksmithing.recipes.extend(RecipeBridge.parse(mine))  # todo: bronze makes 2x for recipe (matCost/2)
blacksmithing.update_matCosts(auction)
blacksmithing.update_craftable_matCosts()

jewelcrafting = RecipeRegistry()
jewelcrafting.recipes = RecipeBridge.parse(jewel)
jewelcrafting.recipes.extend(RecipeBridge.parse(mine))
jewelcrafting.update_matCosts(auction)
jewelcrafting.update_craftable_matCosts()

engineering = RecipeRegistry()
engineering.recipes = RecipeBridge.parse(engin)
engineering.recipes.extend(RecipeBridge.parse(mine))
engineering.update_matCosts(auction)
engineering.update_craftable_matCosts()
```
